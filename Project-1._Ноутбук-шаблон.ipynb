* Неоконченное высшее образование 2000  Балтийская государственная академия рыбопромыслового флота…
Нас будет интересовать только уровень образования.

Создайте с помощью функции-преобразования новый признак **"Образование"**, который должен иметь 4 категории: "высшее", "неоконченное высшее", "среднее специальное" и "среднее".

Выполните преобразование, ответьте на контрольные вопросы и удалите признак "Образование и ВУЗ".

Совет: обратите внимание на структуру текста в столбце **"Образование и ВУЗ"**. Гарантируется, что текущий уровень образования соискателя всегда находится в первых 2ух слов и начинается с заглавной буквы. Воспользуйтесь этим.

```python
#ваш код здесь   
def education_level(data):
    data_list = str(data).split(' ')
    if data_list[0].startswith('В'):
        return data_list[0]+' '+data_list[1]
    elif data_list[0].startswith('Н'):
        return data_list[0]+' '+data_list[1]
    elif data_list[0].startswith('С') and data_list[1].startswith('с'):
        return data_list[0]+" "+data_list[1]+' '+data_list[2]
    else:
        return 'Среднее'

resume_df['Образование'] = resume_df['Образование и ВУЗ'].apply(education_level)
resume_df.drop('Образование и ВУЗ',axis=1,inplace=True)
display(resume_df[resume_df['Образование']=='Среднее'].count())

    
    
        
        
     

  
        

    
        


    
```

3. Следующим этапом преобразуем признак **"Опыт работы"**. Его текущий формат - это: **<Опыт работы: n лет m месяцев, периоды работы в различных компаниях…>**. 

Из столбца нам необходимо выделить общий опыт работы соискателя в месяцах, новый признак назовем "Опыт работы (месяц)"

Для начала обсудим условия решения задачи:
* Во-первых, в данном признаке есть пропуски. Условимся, что если мы встречаем пропуск, оставляем его как есть (функция-преобразование возвращает NaN)
* Во-вторых, в данном признаке есть скрытые пропуски. Для некоторых соискателей в столбце стоит значения "Не указано". Их тоже обозначим как NaN (функция-преобразование возвращает NaN)
* В-третьих, нас не интересует информация, которая описывается после указания опыта работы (периоды работы в различных компаниях)
* В-четвертых, у нас есть проблема: опыт работы может быть представлен только в годах или только месяцах. Например, можно встретить следующие варианты:
    * Опыт работы 3 года 2 месяца…
    * Опыт работы 4 года…
    * Опыт работы 11 месяцев…

В результате преобразования у вас должен получиться столбец, содержащий информацию о том, сколько месяцев проработал соискатель.
Выполните преобразование, ответьте на контрольные вопросы и удалите столбец **"Опыт работы"** из таблицы.


```python
#ваш код здесь
def work_experience(value):
    if value is np.nan or value == 'Не указано':
        return np.nan
    list_value = str(value).split(' ')[:7]
    year = ['год','года','лет']
    month = ['месяц','месяца','месяцев']
    result = 0
    result_1 = 0 
    for x,elem in enumerate(list_value):
        if elem in year:
           result = int(list_value[x-1]) 
        elif elem in month:
           result_1 = int(list_value[x-1])
    return result * 12 + result_1
       
resume_df['Стаж работы'] = resume_df['Опыт работы'].apply(work_experience)
resume_df.drop('Опыт работы',axis=1,inplace=True)
display(resume_df)
display(round(resume_df['Стаж работы'].median()))
```

Выполните преобразования и удалите столбец **"Город, переезд, командировки"** из таблицы.

*Совет: обратите внимание на то, что структура текста может меняться в зависимости от указания ближайшего метро. Учите это, если будете использовать порядок слов в своей программе.*


```python
#ваш код здесь
display(resume_df['Город, переезд, командировки'])

def city(value):
    value = str(value).strip(' ')
    list_value = str(value).split(',')
    million_cities = ['Новосибирск', 'Екатеринбург','Нижний Новгород','Казань', 'Челябинск','Омск', 'Самара', 'Ростов-на-Дону', 'Уфа', 'Красноярск', 'Пермь', 'Воронеж','Волгоград']
    a = list_value[0].rfind(' ')
    if list_value[0][0:a].startswith('Москва'):
        return 'Москва'
    elif list_value[0][0:a].startswith('Санкт-Петербург'):
        return 'Санкт-Петербург'
    for city in million_cities:
        if list_value[0][0:a].startswith(city):
            return 'город-миллионник'
    else:
        return 'другие'
    
resume_df['Город'] = resume_df['Город, переезд, командировки'].apply(city)
display(round(resume_df['Город'].value_counts(normalize=True) * 100))

def removal(value):
    value_list = ['не готова к переезду','не готов к переезду']
    value = str(value).replace(' ','').replace(',','').replace('.','')
    if value.find(value_list[0].replace(' ','')) > 0 or value.find(value_list[1].replace(' ','')) > 0:
        return False
    elif value.find(value_list[0].replace(' ','')) == -1 or value.find(value_list[1].replace(' ','')) == -1:
        return True

resume_df['Переезд'] = resume_df['Город, переезд, командировки'].apply(removal)


def business_trips(value):
    value_list = ['не готова к командировкам','не готов к командировкам']
    value = str(value).replace(' ','').replace(',','').replace('.','')
    if value.find(value_list[0].replace(' ','')) > 0 or value.find(value_list[1].replace(' ','')) > 0:
        return False
    elif value.find(value_list[0].replace(' ','')) == -1 or value.find(value_list[1].replace(' ','')) == -1:
        return True

resume_df['Командировки'] = resume_df['Город, переезд, командировки'].apply(business_trips)
resume_df.drop('Город, переезд, командировки',axis=1,inplace=True)
display(round(resume_df['Город'].value_counts(normalize=True) * 100))
mask = ((resume_df['Командировки'] == True) & (resume_df['Переезд'] == True))
display(round(resume_df[mask].shape[0] / resume_df.shape[0] * 100))
    
    
    
    


    
    
            
            
        
        
```

5. Рассмотрим поближе признаки **"Занятость"** и **"График"**. Сейчас признаки представляют собой набор категорий желаемой занятости (полная занятость, частичная занятость, проектная работа, волонтерство, стажировка) и желаемого графика работы (полный день, сменный график, гибкий график, удаленная работа, вахтовый метод).
На сайте hh.ru соискатель может указывать различные комбинации данных категорий, например:
* полная занятость, частичная занятость
* частичная занятость, проектная работа, волонтерство
* полный день, удаленная работа
* вахтовый метод, гибкий график, удаленная работа, полная занятость

Такой вариант признаков имеет множество различных комбинаций, а значит множество уникальных значений, что мешает анализу. Нужно это исправить!

Давайте создадим признаки-мигалки для каждой категории: если категория присутствует в списке желаемых соискателем, то в столбце на месте строки рассматриваемого соискателя ставится True, иначе - False.

Такой метод преобразования категориальных признаков называется One Hot Encoding и его схема представлена на рисунке ниже:
<img src=https://raw.githubusercontent.com/AndreyRysistov/DatasetsForPandas/main/ohe.jpg>
Выполните данное преобразование для признаков "Занятость" и "График", ответьте на контрольные вопросы, после чего удалите их из таблицы

```python
#ваш код здесь
chart = ['полный день','удаленная работа','вахтовый метод', 'гибкий график', 'сменный график']
busyness = ['проектная работа', 'волонтерство', 'стажировка', 'полная занятость', 'частичная занятость']

for ch,bus in zip(chart,busyness):
    resume_df[ch] = resume_df['График'].apply(lambda x: ch in x)
    resume_df[bus] = resume_df['Занятость'].apply(lambda x: bus in x)
display(resume_df[resume_df['волонтерство'] & resume_df['проектная работа']].shape[0])
display(resume_df[resume_df['вахтовый метод'] & resume_df['гибкий график']].shape[0])

resume_df.drop(['График','Занятость'],axis=1,inplace=True)
display(resume_df)
```

```python
#ваш код здесь
exchangerates_data = pd.read_csv('ExchangeRates.csv',sep=',')
exchangerates_data['date'] = pd.to_datetime(exchangerates_data['date'],dayfirst=True).dt.date
exchangerates_data.drop(['per','time','vol'],axis=1,inplace=True)
#isplay(exchangerates_data)

resume_df['Обновление резюме'] = pd.to_datetime(resume_df['Обновление резюме']).dt.date
resume_df['Зарплата'] = resume_df['ЗП'].apply(lambda x: str(x).split()[0])
resume_df['Зарплата'] = resume_df['Зарплата'].astype('int64')
resume_df['currency'] = resume_df['ЗП'].apply(lambda x: str(x).split()[1])
resume_df['currency'] = resume_df['currency'].replace({'руб.':'RUB', 'бел.руб.':'BYN', 'грн.':'UAH', 'сум':'UZS'})
merge_df = resume_df.merge(exchangerates_data,left_on=['currency','Обновление резюме'],right_on=['currency','date'],how='left')
merge_df['close'].fillna(1,inplace=True)
merge_df['proportion'].fillna(1,inplace=True)
merge_df['ЗП (руб)'] =  merge_df['Зарплата'] * merge_df['close'] / merge_df['proportion']
merge_df.drop(['ЗП','date_x','Зарплата','currency','date_y','close','proportion'],axis=1,inplace=True)
display(round(merge_df['ЗП (руб)'].median()) / 1000)
display(merge_df.info())





```
